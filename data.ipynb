```python
def add_text_and_func(text: str = None, class_name: str = None, dataframe: pd.DataFrame = None) -> pd.DataFrame:
    if text is None or class_name is None:
        print("AttributeError: Some input(s) are missing.")
        return None
    
    text = text.replace('\n', '')
    query_list = text.split('.')
    query_list = query_list[:len(query_list)-1]
    
    new_rows = pd.DataFrame({'String': query_list, 'Class': [class_name] * len(query_list)})
    
    if dataframe is None or dataframe.empty:
        dataframe = new_rows
    else:
        dataframe = pd.concat([dataframe, new_rows], ignore_index=True)
    return dataframe
```
